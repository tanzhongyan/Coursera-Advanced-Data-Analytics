### Data preperation for model usage
In this section, we will create training and testing set with a proportion of 80-20 while stratifying the <code>left</code> class. However, before we start, we need to oversample the minority class as recall is of priority here.

#### Seperating the data into Training and Testing sets (80-20)

```python
X = df2.drop('left',axis=1)
y = df2['left']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.20, random_state=0)
```

```python
#Show class imbalance
y_train.value_counts()
```

#### Oversampling the minority class

```python
smote = SMOTE(random_state=0)
```

```python
X_train, y_train = smote.fit_resample(X_train, y_train)
```

```python
#Proof that there is oversampling of the minority class
y_train.value_counts()
```

### Kicking off the first decision tree model

```python
# Initialise model
decision_tree = DecisionTreeClassifier(random_state=0)

# Fit model to training data
decision_tree.fit(X_train,y_train)
```

# pacE: Execute Stage
- Interpret model performance and results
- Share actionable steps with stakeholders



```python
# Make predictions on test data
dt_pred = decision_tree.predict(X_test)

# Generate performance metrics
print("Accuracy:", "%.3f" % accuracy_score(y_test, dt_pred))
print("Precision:", "%.3f" % precision_score(y_test, dt_pred))
print("Recall:", "%.3f" % recall_score(y_test, dt_pred))
print("F1 Score:", "%.3f" % f1_score(y_test, dt_pred))
```

```python
#Building Confusion matrix
cm = confusion_matrix(y_test, dt_pred, labels=decision_tree.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels = decision_tree.classes_)
disp.plot(values_format='')
plt.title('Confusion Matrix for Decision Tree')
plt.show()
```

```python
# Plot the tree
plt.figure(figsize=(50,25))
plot_tree(decision_tree, max_depth=4, fontsize=14, feature_names=X.columns, 
          class_names={0:'stayed', 1:'left'}, filled=True);
plt.show()
```
